```python
import pandas as pd
```

```python
df = pd.read_csv('Advertising.csv')
```

```python
df
```

```python
X = df[['TV', 'Radio', 'Newspaper']]
```

```python
y = df.Sales
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(X_train, y_train)
```

```python
predictions = lin_reg.predict(X_test)
```

```python
print('Intercept:', lin_reg.intercept_)
print('Coefficients:', lin_reg.coef_)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
print('Mean Absolute Error (MAE):', mean_absolute_error(y_test,predictions))
```

```python
print('Mean Squared Error(MSE):', mean_squared_error(y_test, predictions))
```

```python
print('Root Mean Squared Error (RMSE):', np.sqrt(mean_squared_error(y_test, predictions)))
```

```python
print('R-Square:', r2_score(y_test, predictions))
```

```python
fig, axs = plt.subplots(3, figsize=(10,8), sharex=True, dpi=600)
axs[0].scatter(X_test.TV,y_test, label='Actual Sales TV' )
axs[0].scatter(X_test.TV, predictions, label='Predicted Sales TV')
axs[0].legend(loc=1)
axs[1].scatter(X_test.Radio, y_test, label='Actual Sales Radio')
axs[1].scatter(X_test.Radio, predictions, label='Predicted Sales Radio')
axs[1].legend()
axs[2].scatter(X_test.Newspaper, y_test, label='Actual Sales Newspaper')
axs[2].scatter(X_test.Newspaper, predictions, label='Predicted Sales Newspaper')
axs[2].legend()
plt.show()
```

```python
import pandas as pd
```

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
feature_set = ['pregnant', 'insulin', 'bmi', 'age', 'glucose', 'bp', 'pedigree']
```

```python
features = diabetes[feature_set]
```

```python
target = diabetes.label
```

```python
from sklearn.model_selection import train_test_split
```

```python
feature_train, feature_test, target_train, target_test = train_test_split(features, target, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
logreg = LogisticRegression(solver='lbfgs')
```

```python
logreg.fit(feature_train, target_train)
```

```python
predictions = logreg.predict(feature_test)
```

```python
print('Logistic Regression Model Accuracy:', accuracy_score(target_test, predictions))
```
